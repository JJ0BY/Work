```python
#Create modules 
from qiskit import QuantumCircuit, Aer, assemble
from qiskit.circuit import Gate
import numpy as np
from math import pi
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_state_qsphere
```

```python
#We are showing that different gates can be used to create other gates 
"""
HZH = X
HXH = Z
SX(1/S) = Y
SY(1/S) = X
SZ(1/S) = Z
ZZZ = Z
ZXZ = -X
ZYZ = -Y
"""

qc = QuantumCircuit(2)
c = 0
t = 1
qc.cz(c,t)
qc.draw(output='mpl')

```

```python
#We are creating a multi qubit gate such that, if an X gate is applied on q1 into a Z gate, thus controlled Z gate 

qc = QuantumCircuit(2)
# also a controlled-Z
qc.h(t)
qc.cx(c,t)
qc.h(t)
qc.draw(output='mpl')

```

```python
#Controlled Y gate

qc = QuantumCircuit(2)
# a controlled-Y
qc.sdg(t)
qc.cx(c,t)
qc.s(t)
qc.draw(output='mpl')
```

```python
#Controlled H gate

qc = QuantumCircuit(2)
# a controlled-H
qc.ry(pi/4,t)
qc.cx(c,t)
qc.ry(-pi/4,t)
qc.draw(output='mpl')
```

```python
a = 0
b = 1
qc = QuantumCircuit(2)
# swaps states of qubits a and b
qc.swap(a,b)
qc.draw(output='mpl')

#The swap gate are 2 CNOT gates applied to each other, with each one in a different direction 
```

```python
#Controlled rotations 

qc = QuantumCircuit(2)
theta = pi # theta can be anything (pi chosen arbitrarily)
qc.ry(theta/2,t)
qc.cx(c,t)
qc.ry(-theta/2,t)
qc.cx(c,t)
qc.draw(output='mpl')
```

```python
"""
Using 3 matrices ABC and a scalar 'a', one can create any gate V, where

ABC = I, the identity Matrix 

e^(ia)AZBZCZ = V 

"""


A = Gate('A', 1, [])
B = Gate('B', 1, [])
C = Gate('C', 1, [])
alpha = 1 # arbitrarily define alpha to allow drawing of circuit

qc = QuantumCircuit(2)
qc.append(C, [t])
qc.cz(c,t)
qc.append(B, [t])
qc.cz(c,t)
qc.append(A, [t])
qc.p(alpha,c)
qc.draw(output='mpl')
```

```python
#Toffoli gate is similar to the AND gate

#Both q0 and q1 should be |1>, and then q2 = |1> 
qc = QuantumCircuit(3)
a = 0
b = 1
t = 2
# Toffoli with control qubits a and b and target t
qc.ccx(a,b,t)
qc.draw(output='mpl')
```

```python
#A Toffoli gate made of 2 control H gates and a CNOT gate 
qc = QuantumCircuit(3)
qc.ch(a,t)
qc.cz(b,t)
qc.ch(a,t)
qc.draw(output='mpl')
```

```python
#Errors can occur while changing qubits 
#In this example, we are rotating the vector in a way assuming the H and T gate are close to perfect
#This helps minimize error 
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.draw(output='mpl')

# Let's see the result
svsim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Statevector} = "))
plot_bloch_multivector(final_state)
```

```python
#Using multiple H and T gates in succession can rotate the arrow with minimal error 
qc = QuantumCircuit(1)
qc.h(0)
qc.t(0)
qc.h(0)
qc.t(0)
qc.draw(output='mpl')

# Let's see the result
svsim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
final_state = svsim.run(qobj).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Statevector} = "))
plot_bloch_multivector(final_state)
```
