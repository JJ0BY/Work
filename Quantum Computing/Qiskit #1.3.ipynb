```python
#import modules 
from qiskit import QuantumCircuit, assemble, Aer
from math import pi, sqrt
from qiskit.visualization import plot_bloch_multivector, plot_histogram
```

```python
#Create quantum computer simulator
sim = Aer.get_backend('aer_simulator')
```

```python
# Control bloch sphere
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Pauli-X gate (NOT gate)

The pauli X matrix is 

X = [ 0  1 ]
    [ 1  0 ]  

format: constant|vector> = [element1 element2], where the constant can be a complex number 

When X is matrix multiplied by  i|1> = [ i  0 ], result is   i|0> = [ 0  i ]
When X is matrix multiplied by   |1> = [ 1  0 ], result is    |0> = [ 0  1 ]
When X is matrix multiplied by  -|1> = [-1  0 ], result is   -|0> = [ 0 -1 ]
When X is matrix multiplied by -i|1> = [-i  0 ], result is  -i|0> = [ 0 -i ]

When X is matrix multiplied by  i|0> = [ 0  i ], result is   i|1> = [ i  0 ]
When X is matrix multiplied by   |0> = [ 0  1 ], result is    |1> = [ 1  0 ]
When X is matrix multiplied by  -|0> = [ 0 -1 ], result is   -|1> = [-1  0 ]
When X is matrix multiplied by -i|0> = [ 0 -i ], result is  -i|1> = [-i  0 ]

"""

# Let's do an X-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.x(0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Pauli-Y gate 

The pauli Y matrix is 

Y = [ 0 -i ]
    [ i  0 ]
    
format: constant|vector> = [element1 element2], where the constant can be a complex number 

When X is matrix multiplied by  i|1> = [ i  0 ], result is   -|0> = [ 0 -1 ]
When X is matrix multiplied by   |1> = [ 1  0 ], result is   i|0> = [ 0  i ]
When X is matrix multiplied by  -|1> = [-1  0 ], result is  -i|0> = [ 0 -i ]
When X is matrix multiplied by -i|1> = [-i  0 ], result is    |0> = [ 0  1 ]

When X is matrix multiplied by  i|0> = [ 0  i ], result is    |1> = [ 1  0 ]
When X is matrix multiplied by   |0> = [ 0  1 ], result is  -i|1> = [-i  0 ]
When X is matrix multiplied by  -|0> = [ 0 -1 ], result is   i|1> = [ i  0 ]
When X is matrix multiplied by -i|0> = [ 0 -i ], result is   -|1> = [-1  0 ]

"""




# Let's do an Y-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.y(0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Pauli-Z gate 

The pauli Z matrix is 

Y = [ 1  0 ]
    [ 0 -1 ]
    
format: constant|vector> = [element1 element2], where the constant can be a complex number 

When X is matrix multiplied by  i|1> = [ i  0 ], result is   i|1> = [ i  0 ]
When X is matrix multiplied by   |1> = [ 1  0 ], result is    |1> = [ 1  0 ]
When X is matrix multiplied by  -|1> = [-1  0 ], result is   -|1> = [-1  0 ]
When X is matrix multiplied by -i|1> = [-i  0 ], result is  -i|1> = [-i  0 ]

When X is matrix multiplied by  i|0> = [ 0  i ], result is  -i|0> = [ 0 -i ]
When X is matrix multiplied by   |0> = [ 0  1 ], result is   -|0> = [ 0 -1 ]
When X is matrix multiplied by  -|0> = [ 0 -1 ], result is    |0> = [ 0  1 ]
When X is matrix multiplied by -i|0> = [ 0 -i ], result is   i|0> = [ 0  i ]

"""




# Let's do an Z-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.z(0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Hadamard gate 

The hadamard matrix is 

Y =  [ 1/sqrt(2)  1/sqrt(2) ]
     [ 1/sqrt(2) -1/sqrt(2) ]
    
format: constant|vector> = [element1 element2], where the constant can be a complex number 

When X is matrix multiplied by  i|1> = [ i  0 ], result is  1/sqrt(2)[ i  i ]
When X is matrix multiplied by   |1> = [ 1  0 ], result is  1/sqrt(2)[ 1  1 ]
When X is matrix multiplied by  -|1> = [-1  0 ], result is  1/sqrt(2)[-1 -1 ]
When X is matrix multiplied by -i|1> = [-i  0 ], result is  1/sqrt(2)[-i -i ]

When X is matrix multiplied by  i|0> = [ 0  i ], result is  1/sqrt(2)[ i -i ]
When X is matrix multiplied by   |0> = [ 0  1 ], result is  1/sqrt(2)[ 1 -1 ]
When X is matrix multiplied by  -|0> = [ 0 -1 ], result is  1/sqrt(2)[-1  1 ]
When X is matrix multiplied by -i|0> = [ 0 -i ], result is  1/sqrt(2)[-i  i ]

"""




# Let's do an H-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.h(0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Identity gate 

The identity matrix is 

I =  [ 1  0 ]
     [ 0  1 ]

The vector or matrix times the identitiy matrix, is the original matric itself 
"""

# Let's do an H-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.i(0)
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Special Phase gate 

The Special phase matrix is

S =  [ 1     0        ]
     [ 0  e^(i(pi/2)) ]

The S gate is a P gate with c = pi/2. It is special because every other gate is it's own inverse. 

The inverse of the Special phase matrix is

1/S =  [ 1      0        ]
       [ 0  e^(-i(pi/2)) ]

SS = Z
S(1/S) = I

"""

# Let's do an H-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)
qc.s(0)   # Apply S-gate to qubit 0
qc.sdg(0) # Apply Sdg-gate to qubit 0
qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

```python
"""
Transpose gate 

The Transpose phase matrix is

T =  [ 1     0        ]
     [ 0  e^(i(pi/4)) ]

The T gate is a P gate with c = pi/2. Also has a different inverse

The inverse of the Transpose phase matrix is

1/T =  [ 1      0        ]
       [ 0  e^(-i(pi/4)) ]

TTTT = Z
T(1/T) = I

"""

# Let's do an H-gate on a |0> qubit
qc = QuantumCircuit(1)
initial_state = [-1j/sqrt(3), sqrt(2)/sqrt(3)]
qc.initialize(initial_state, 0)

#TTTTT(1/T) = Z
qc.t(0)   # Apply S-gate to qubit 0
qc.t(0)   # Apply S-gate to qubit 0
qc.t(0)   # Apply S-gate to qubit 0
qc.t(0)   # Apply S-gate to qubit 0
qc.t(0)   # Apply S-gate to qubit 0
qc.tdg(0) # Apply Sdg-gate to qubit 0


qc.draw(output='mpl')

# Let's see the result
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```
