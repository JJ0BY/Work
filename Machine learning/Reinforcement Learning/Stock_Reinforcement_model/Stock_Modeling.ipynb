```python
import gym
import gym_anytrading

from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3 import A2C

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt 

from gym_anytrading.envs import StocksEnv
from finta import TA
```

```python
df = pd.read_csv('MSFTdata.csv')
```

```python
df.head()
print(len(df))
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df.sort_values('Date', ascending=True, inplace=True)
df.head()
```

```python
df.set_index('Date', inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df['Volume'] = df['Volume'].astype(float)
df.dtypes
```

```python
df['SMA'] = TA.SMA(df, 12)
df['RSI'] = TA.RSI(df)
df['OBV'] = TA.OBV(df)
df.fillna(0, inplace=True)
```

```python
df.head(15)
```

```python
env = gym.make('stocks-v0', df=df, frame_bound=(10,5000), window_size=5)
```

```python
def add_signals(env):
    start = env.frame_bound[0] - env.window_size
    end = env.frame_bound[1]
    prices = env.df.loc[:, 'Close'].to_numpy()[start:end]
    signal_features = env.df.loc[:, ['Close', 'Volume', 'SMA', 'RSI', 'OBV']].to_numpy()[start:end]
    
    return prices, signal_features
    
```

```python
class myCustomEnv(StocksEnv):
    
    _process_data = add_signals 
    
env2 = myCustomEnv(df=df, window_size = 100, frame_bound=(7000,8000))
```

```python
#env_maker = lambda: gym.make('stocks-v0', df=df, frame_bound=(5,100), window_size=5)
env_maker = lambda: env2
env = DummyVecEnv([env_maker])
```

```python
model = A2C('MlpPolicy', env, verbose=1)
model.learn(total_timesteps=100000)
```

```python
env = myCustomEnv(df=df, window_size = 100, frame_bound=(9000,9100))
obs = env.reset()

while True:
    obs = obs[np.newaxis, ...]
    
    action, _states = model.predict(obs)
    obs, reward, done, info = env.step(action)
    
    if done:
        print('info', info)
        break
    
plt.figure(figsize =(15,6))
plt.cla()
env.render_all()
plt.show()
```
