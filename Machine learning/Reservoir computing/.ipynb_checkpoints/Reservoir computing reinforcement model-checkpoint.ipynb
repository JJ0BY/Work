```python
import rctorch as r
import torch as t
import os
from reservoirpy.datasets import mackey_glass
from reservoirpy.observables import rmse, nrmse, rsquare
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler as MMS
from reservoirpy.hyper import research, plot_hyperopt_report
from reservoirpy.observables import rmse, nrmse, rsquare
import warnings 
import time
warnings.filterwarnings('ignore')
```

```python
def MSE(y_test, y_pred, PRINT=True):
    
    length = len(y_test)
    error = rmse(y_test, y_pred)
    r2 = rsquare(y_test, y_pred)
    
    r2 = 1-((1-r2)*(length-1)/(length-7-1))
    
    if PRINT:
        
        print(f'Root Mean Squared Error: {error}')
        print(f'Average Root Mean Squared Error: {error/length}\n')
        print(f'Adjusted R^2 error: {r2}')
    
    return r2, error
```

```python
Steps = 2000

X = mackey_glass(Steps).astype(np.float32)

plt.figure(figsize=(10, 3))
plt.title(f"Mackey Glass with {Steps} steps")
plt.ylabel("$f(x)$")
plt.xlabel("$x$")
plt.plot(X)
plt.show()
```

```python
begin = 0
start = 1000
add = 999

Scaler = MMS()
X = Scaler.fit_transform(X)
    
x_train = t.tensor(X[begin:begin+start])
y_train = t.tensor(X[begin+1:begin+start+1])
x_test = t.tensor(X[begin+start:begin+start+add])
y_test = t.tensor(X[begin+start+1:begin+start+add+1])

norm_test = Scaler.inverse_transform(y_test)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
hyps = {
    
    'connectivity':0.5,
    'spectral_radius':0.5,
    'n_nodes':100,
    'regularization': 0.01,
    'leaking_rate': 0.01,
    'bias': 0.5
    
}
```

```python
t1 = time.time()
model = r.RcNetwork(**hyps, random_state = 1234, feedback = True)

model.fit(x_train, y_train)

score, pred = model.test(x_test, y_test)
t2 = time.time()
print(f'{t2-t1} s')
```

```python
# Objective functions accepted by ReservoirPy must respect some conventions:
#  - dataset and config arguments are mandatory, like the empty '*' expression.
#  - all parameters that will be used during the search must be placed after the *.
#  - the function must return a dict with at least a 'loss' key containing the result
# of the loss function. You can add any additional metrics or information with other
# keys in the dict. See hyperopt documentation for more informations.
def objective(dataset, config, *, N, sr, lr, iss, c, b, reg):

    # This step may vary depending on what you put inside 'dataset'
    train_data, validation_data = dataset
    X_train, y_train = train_data
    X_test, Y_test = validation_data

    # You can access anything you put in the config
    # file from the 'config' parameter.
    instances = config["instances_per_trial"]

    # The seed should be changed across the instances,
    # to be sure there is no bias in the results
    # due to initialization.
    variable_seed = 1234

    losses = []; r2s = [];
    
    hyps = {
    
    'connectivity':c,
    'spectral_radius':sr,
    'n_nodes':N,
    'regularization': reg,
    'leaking_rate': lr,
    'bias': b,
    'input_scaling': iss
    
    }
    
    for n in range(instances):
        # Build your model given the input parameters
        model = r.RcNetwork(**hyps, random_state = variable_seed, feedback = True)

        model.fit(x_train, y_train)
        
        # Train your model and test your model.
        
        score, predictions= model.test(x_test, y_test)
        
        loss = nrmse(Y_test, predictions, norm_value=np.ptp(X_train))
        r2 = rsquare(Y_test, predictions)

        # Change the seed between instances
        variable_seed += 1

        losses.append(loss)
        r2s.append(r2)

    # Return a dictionnary of metrics. The 'loss' key is mandatory when
    # using hyperopt.
    loss = float(np.mean(losses))
    r2 = float(np.mean(losses))
    
    dic = {'loss': loss, 'r2': r2}

    return dic
```

```python
dataset = ((x_train, y_train), (x_test, y_test))
```

```python
hyperopt_config = {
    "exp": f"Mackey Glass", # the experimentation name
    "hp_max_evals": 250,             # the number of differents sets of parameters hyperopt has to try
    "hp_method": "random",           # the method used by hyperopt to chose those sets (see below)
    "seed": 42,                      # the random state seed, to ensure reproducibility
    "instances_per_trial": 3,        # how many random ESN will be tried with each sets of parameters
    "hp_space": {                    # what are the ranges of parameters explored
        "N": ["choice", 100],             # the number of neurons is fixed to 300
        "sr": ["loguniform", 1, 10],   # the spectral radius is log-uniformly distributed between 1e-6 and 10
        "lr": ["loguniform", 1e-3, 1],  # idem with the leaking rate, from 1e-3 to 1
        "iss": ["choice", 1],           # the input scaling is fixed
        "reg": ["loguniform", 0.1, 1],        # and so is the regularization parameter.
        "c": ["loguniform", 0.1, 1],
        "b": ["loguniform", 0.1, 1],
        
    }
}

# we precautionously save the configuration in a JSON file
# each file will begin with a number corresponding to the current experimentation run number.
with open(f"{hyperopt_config['exp']}.config.json", "w+") as f:
    json.dump(hyperopt_config, f)
```
