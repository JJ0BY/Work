```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random as r
```

```python
def f(x):
    
    num = x**3+5*x**2+x*3
    
    random = r.random()
    sign = r.randint(0,1)
    if sign == 0:
        num += random
    else:
        num -= random
        
    return num

def graph(L, X1, Y1):
    
    X = []
    Y = []
    
    for i in range(len(L)):
        
        X.append(L[i][0])
        Y.append(L[i][1])
        
    plt.plot(X,Y, 'o')
    
    plt.plot(X1,Y1, 'o')
        
```

```python
a = -5
b = 1

n = 1000

L = []
for i in range (n):
    
    num = r.randint(a, b)
    sign = r.randint(0,1)
    
    random = r.random()
    
    if sign == 0:
        num += random
    else:
        num -= random
        
    ans = f(num)
    
    sign = r.randint(0,1)
    
    random = r.random()
    if sign == 0:
        num += random
    else:
        num -= random
        
    L.append([num, ans])
    
    X = np.linspace(a, b, n)
    Y = f(X)
```

```python
graph(L, X, Y)
```
